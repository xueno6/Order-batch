```python
import time, datetime
import numpy as np
from gurobipy import *
def Return_CWII_Heuristic_Solver(s):
    start=time.time()
    def save_fun(s,L1,L2):
        return Obj_Matrix_Return_solver(s,[L1])+Obj_Matrix_Return_solver(s,[L2])-Obj_Matrix_Return_solver(s,[L1+L2])
    def delete_used_orders(s,batch,d):
        record=[]
        for i in d.keys():
            if i[0] in batch:
                record.append(i)
            elif i[1] in batch:
                record.append(i)
            else:
                pass
        for j in record:
            d.pop(j)
        return d
    %run Obj_Matrix_Return.ipynb


    #save_matrix=np.zeros((s.number_of_order+1,s.number_of_order))
    CWII_matrix=[[] for i in range(s.number_of_batch)]
    number_per_unit=int(s.number_of_order/s.number_of_batch)
    dict_index=[]
    dict_value=[]

    for i in range(s.number_of_order+1):
        if i < s.number_of_order:
            for j in range(i+1,s.number_of_order):
                dict_index.append((i,j))
                dict_value.append(save_fun(s,[i],[j]))
        else:
            for j in range(0,s.number_of_order):
                dict_index.append((i,j))
                dict_value.append(0)
    d = tupledict(zip(dict_index,dict_value))
    d=tupledict([(k,v) for k, v in sorted(d.items(), key=lambda item: item[1], reverse=True)])
    index=0
    order_dict_index=[i for i in range(s.number_of_order)]
    order_dict_value=[0 for i in range(s.number_of_order)]
    order_dict=tupledict(zip(order_dict_index,order_dict_value))
    while index<s.number_of_batch:
        New_order_add=0
        i,j=d.keys()[0]
        d.pop((i,j))
        if i<s.number_of_order:
            if (i not in CWII_matrix[index]) and (j not in CWII_matrix[index]):
                if number_per_unit-len(CWII_matrix[index])>=2:
                    CWII_matrix[index].extend([i,j])
                    order_dict.pop(i)
                    order_dict.pop(j)
                    New_order_add=1
                    if number_per_unit-len(CWII_matrix[index])==0:
                        d=delete_used_orders(s,CWII_matrix[index],d)
                        index=index+1

            elif (i in CWII_matrix[index]) and (j not in CWII_matrix[index]):
                if number_per_unit-len(CWII_matrix[index])>=1:
                    CWII_matrix[index].append(j)
                    order_dict.pop(j)
                    New_order_add=1
                    if number_per_unit-len(CWII_matrix[index])==0:
                        d=delete_used_orders(s,CWII_matrix[index],d)
                        index=index+1

            elif (j in CWII_matrix[index]) and (i not in CWII_matrix[index]):
                if number_per_unit-len(CWII_matrix[index])>=1:
                    CWII_matrix[index].append(i)
                    order_dict.pop(i)
                    New_order_add=1
                    if number_per_unit-len(CWII_matrix[index])==0:
                        d=delete_used_orders(s,CWII_matrix[index],d)
                        index=index+1
            else:
                pass
        else:
            if (j not in CWII_matrix[index]) and number_per_unit-len(CWII_matrix[index])>=1:
                CWII_matrix[index].append(j)
                order_dict.pop(j)
                New_order_add=1
                if number_per_unit-len(CWII_matrix[index])==0:
                    d=delete_used_orders(s,CWII_matrix[index],d)
                    index=index+1

        if index!=s.number_of_batch and len(CWII_matrix[index])>0 and New_order_add==1:
            for j in range(s.number_of_order):
                #if (d.select(j,'*')+d.select('*',j)!=[]) and j not in CWII_matrix[index]:
                if j in order_dict.keys():
                    try:
                        d[(s.number_of_order,j)]=save_fun(s,CWII_matrix[index],[j])
                    except:
                        pass
            d=tupledict([(k,v) for k, v in sorted(d.items(), key=lambda item: item[1], reverse=True)])       
    obj=Obj_Matrix_Return_solver(s,CWII_matrix)
    end=time.time()
    return CWII_matrix,obj,end-start
        
        
        
        

def Sshape_CWII_Heuristic_Solver(s):
    start=time.time()
    def save_fun(s,L1,L2):
        return Obj_Matrix_Sshape_solver(s,[L1])+Obj_Matrix_Sshape_solver(s,[L2])-Obj_Matrix_Sshape_solver(s,[L1+L2])
    def delete_used_orders(s,batch,d):
        record=[]
        for i in d.keys():
            if i[0] in batch:
                record.append(i)
            elif i[1] in batch:
                record.append(i)
            else:
                pass
        for j in record:
            d.pop(j)
        return d
    %run Obj_Matrix_Return.ipynb


    #save_matrix=np.zeros((s.number_of_order+1,s.number_of_order))
    CWII_matrix=[[] for i in range(s.number_of_batch)]
    number_per_unit=int(s.number_of_order/s.number_of_batch)
    dict_index=[]
    dict_value=[]

    for i in range(s.number_of_order+1):
        if i < s.number_of_order:
            for j in range(i+1,s.number_of_order):
                dict_index.append((i,j))
                dict_value.append(save_fun(s,[i],[j]))
        else:
            for j in range(0,s.number_of_order):
                dict_index.append((i,j))
                dict_value.append(0)
    d = tupledict(zip(dict_index,dict_value))
    d=tupledict([(k,v) for k, v in sorted(d.items(), key=lambda item: item[1], reverse=True)])
    index=0
    order_dict_index=[i for i in range(s.number_of_order)]
    order_dict_value=[0 for i in range(s.number_of_order)]
    order_dict=tupledict(zip(order_dict_index,order_dict_value))
    while index<s.number_of_batch:
        New_order_add=0
        i,j=d.keys()[0]
        d.pop((i,j))
        if i<s.number_of_order:
            if (i not in CWII_matrix[index]) and (j not in CWII_matrix[index]):
                if number_per_unit-len(CWII_matrix[index])>=2:
                    CWII_matrix[index].extend([i,j])
                    order_dict.pop(i)
                    order_dict.pop(j)
                    New_order_add=1
                    if number_per_unit-len(CWII_matrix[index])==0:
                        d=delete_used_orders(s,CWII_matrix[index],d)
                        index=index+1

            elif (i in CWII_matrix[index]) and (j not in CWII_matrix[index]):
                if number_per_unit-len(CWII_matrix[index])>=1:
                    CWII_matrix[index].append(j)
                    order_dict.pop(j)
                    New_order_add=1
                    if number_per_unit-len(CWII_matrix[index])==0:
                        d=delete_used_orders(s,CWII_matrix[index],d)
                        index=index+1

            elif (j in CWII_matrix[index]) and (i not in CWII_matrix[index]):
                if number_per_unit-len(CWII_matrix[index])>=1:
                    CWII_matrix[index].append(i)
                    order_dict.pop(i)
                    New_order_add=1
                    if number_per_unit-len(CWII_matrix[index])==0:
                        d=delete_used_orders(s,CWII_matrix[index],d)
                        index=index+1
            else:
                pass
        else:
            if (j not in CWII_matrix[index]) and number_per_unit-len(CWII_matrix[index])>=1:
                CWII_matrix[index].append(j)
                order_dict.pop(j)
                New_order_add=1
                if number_per_unit-len(CWII_matrix[index])==0:
                    d=delete_used_orders(s,CWII_matrix[index],d)
                    index=index+1

        if index!=s.number_of_batch and len(CWII_matrix[index])>0 and New_order_add==1:
            for j in range(s.number_of_order):
                #if (d.select(j,'*')+d.select('*',j)!=[]) and j not in CWII_matrix[index]:
                if j in order_dict.keys():
                    try:
                        d[(s.number_of_order,j)]=save_fun(s,CWII_matrix[index],[j])
                    except:
                        pass
            d=tupledict([(k,v) for k, v in sorted(d.items(), key=lambda item: item[1], reverse=True)])   
    obj=Obj_Matrix_Sshape_solver(s,CWII_matrix)
    end=time.time()
    return CWII_matrix,obj,end-start
```
