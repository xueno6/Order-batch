```python
import time
import numpy as np
from gurobipy import *
def Return_CWI_Heuristic_Solver(s):
    start=time.time()
    def save_fun(s,L1,L2):
        return Obj_Matrix_Return_solver(s,[L1])+Obj_Matrix_Return_solver(s,[L2])-Obj_Matrix_Return_solver(s,[L1+L2])
    def delete_used_orders(batch,d):
        record=[]
        for i in d.keys():
            if i[0] in batch:
                record.append(i)
            elif i[1] in batch:
                record.append(i)
            else:
                pass
        for j in record:
            d.pop(j)
        return d


    %run Obj_Matrix_Return.ipynb


    #save_matrix=np.zeros((s.number_of_order,s.number_of_order))
    CWI_matrix=[[] for i in range(s.number_of_batch)]
    number_per_unit=int(s.number_of_order/s.number_of_batch)

    dict_index=[]
    dict_value=[]
    for i in range(s.number_of_order):
        for j in range(i+1,s.number_of_order):
            dict_index.append((i,j))
            dict_value.append(save_fun(s,[i],[j]))
    d = tupledict(zip(dict_index,dict_value))
    index=0
    d=tupledict([(k,v) for k, v in sorted(d.items(), key=lambda item: item[1], reverse=True)])

    while index<s.number_of_batch:
        i,j=d.keys()[0]
        d.pop((i,j))
        if (i not in CWI_matrix[index]) and (j not in CWI_matrix[index]):
            if number_per_unit-len(CWI_matrix[index])>=2:
                CWI_matrix[index].extend([i,j])
                if number_per_unit-len(CWI_matrix[index])==0:
                    d=delete_used_orders(CWI_matrix[index],d)
                    index=index+1

        elif (i in CWI_matrix[index]) and (j not in CWI_matrix[index]):
            if number_per_unit-len(CWI_matrix[index])>=1:
                CWI_matrix[index].append(j)
                if number_per_unit-len(CWI_matrix[index])==0:
                    d=delete_used_orders(CWI_matrix[index],d)
                    index=index+1

        elif (j in CWI_matrix[index]) and (i not in CWI_matrix[index]):
            if number_per_unit-len(CWI_matrix[index])>=1:
                CWI_matrix[index].append(i)
                if number_per_unit-len(CWI_matrix[index])==0:
                    d=delete_used_orders(CWI_matrix[index],d)
                    index=index+1
        else:
            pass

    obj=Obj_Matrix_Return_solver(s,CWI_matrix)
    end=time.time()
    return CWI_matrix,obj,end-start
        
        
        
        

def Sshape_CWI_Heuristic_Solver(s):
    start=time.time()
    def save_fun(s,L1,L2):
        return Obj_Matrix_Sshape_solver(s,[L1])+Obj_Matrix_Sshape_solver(s,[L2])-Obj_Matrix_Sshape_solver(s,[L1+L2])

    def delete_used_orders(batch,d):
        record=[]
        for i in d.keys():
            if i[0] in batch:
                record.append(i)
            elif i[1] in batch:
                record.append(i)
            else:
                pass
        for j in record:
            d.pop(j)
        return d
    %run Obj_Matrix_Sshape.ipynb
    #save_matrix=np.zeros((s.number_of_order,s.number_of_order))
    CWI_matrix=[[] for i in range(s.number_of_batch)]
    number_per_unit=int(s.number_of_order/s.number_of_batch)

    dict_index=[]
    dict_value=[]
    for i in range(s.number_of_order):
        for j in range(i+1,s.number_of_order):
            dict_index.append((i,j))
            dict_value.append(save_fun(s,[i],[j]))
    d = tupledict(zip(dict_index,dict_value))
    index=0
    d=tupledict([(k,v) for k, v in sorted(d.items(), key=lambda item: item[1], reverse=True)])

    while index<s.number_of_batch and len(d)>0:
        i,j=d.keys()[0]
        d.pop((i,j))
        if (i not in CWI_matrix[index]) and (j not in CWI_matrix[index]):
            if number_per_unit-len(CWI_matrix[index])>=2:
                CWI_matrix[index].extend([i,j])
                if number_per_unit-len(CWI_matrix[index])==0:
                    d=delete_used_orders(CWI_matrix[index],d)
                    index=index+1

        elif (i in CWI_matrix[index]) and (j not in CWI_matrix[index]):
            if number_per_unit-len(CWI_matrix[index])>=1:
                    CWI_matrix[index].append(j)
                    if number_per_unit-len(CWI_matrix[index])==0:
                        d=delete_used_orders(CWI_matrix[index],d)
                        index=index+1

        elif (j in CWI_matrix[index]) and (i not in CWI_matrix[index]):
            if number_per_unit-len(CWI_matrix[index])>=1:
                CWI_matrix[index].append(i)
                if number_per_unit-len(CWI_matrix[index])==0:
                    d=delete_used_orders(CWI_matrix[index],d)
                    index=index+1
        else:
            pass
                
    obj=Obj_Matrix_Sshape_solver(s,CWI_matrix)
    end=time.time() 
    return CWI_matrix,obj,end-start
```
